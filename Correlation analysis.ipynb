```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5,60,10)
y =  [38,33,32,14,20,5]
```

```python
fig,ax = plt.subplots()
ax.scatter(x,y)
ax.set_xlabel('x',fontsize = 16)
ax.set_ylabel('y',fontsize =16)
plt.show()
```

```python
np.corrcoef(x,y)
```

```python
mpg = pd.read_csv('auto-mpg.csv')
```

```python
mpg.head()
```

```python
fig, ax = plt.subplots(nrows=1,ncols= 3,figsize =(13,3))
#scaterplot horsepower
ax[0].scatter(mpg['horsepower'],mpg['mpg'])
ax[0].set_xlabel('horsepower')
ax[0].set_ylabel('mpg')

#scatter plot weight
ax[1].scatter(mpg['weight'],mpg['mpg'])
ax[1].set_xlabel('weight')

#scatter plot acceleration
ax[2].scatter(mpg['acceleration'],mpg['mpg'])
ax[2].set_xlabel('acceleration')
plt.show()
```

```python
mpg.corr(numeric_only=True)
```

```python
corr_hordepower = mpg.corr(numeric_only=True).loc['mpg','horsepower']
corr_weight = mpg.corr(numeric_only=True).loc['mpg','weight']
corr_acceleration = mpg.corr(numeric_only=True).loc['mpg','acceleration']
```

```python
fig,ax =plt.subplots(nrows =1 ,ncols =3,figsize = (13,3))

#scaterplot horsepower
ax[0].scatter(mpg['horsepower'], mpg['mpg'])
ax[0].set_xlabel('Horsepower')
ax[0].set_ylabel('MPG')
ax[0].set_title('r = ' + str(round(corr_horsepower, 2)))

#scatter plot weight
ax[1].scatter(mpg['weight'],mpg['mpg'])
ax[1].set_xlabel('weight')
ax[1].set_title('r = ' + str(round(corr_weight,2)))

#scatter plot acceleration
ax[2].scatter(mpg['acceleration'],mpg['mpg'])
ax[2].set_xlabel('acceleration')
ax[2].set_title('r = ' + str(round(corr_acceleration,2)))
plt.show()
```
